# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def son_hiz(vi,a,t):
    vf = vi + (a * t)
    return vf
son_hiz(24,5,15)
````

# Problem 2

1'den 1000'e kadar olan sayılardan mükemmel sayı olanları ekrana yazdırın. Bunun için bir sayının mükemmel olup olmadığını dönen bir tane fonksiyon yazın.

Bir sayının bölenlerinin toplamı kendine eşitse bu sayı mükemmel bir sayıdır. Örnek olarak 6 mükemmel bir sayıdır (1 + 2 + 3 = 6).

````python
# 2.Problemin Çözümünü Buraya Yazınız
def mukemmel_sayi():
    mukemmel_sayilar = []
    for sayi in range(1,1001):
        toplam = 0
        for i in range(1,sayi):
            a = sayi % i
            if a == 0:
                toplam += i
        if toplam == sayi:
            mukemmel_sayilar.append(sayi)
    return mukemmel_sayilar
mukemmel_sayi()    
````

# Problem 3

1'den 100'e kadar olan sayılardan pisagor üçgeni oluşturanları ekrana yazdıran bir fonksiyon yazın.(a <= 100,b <= 100)


````python
# 3.Problemin Çözümünü Buraya Yazınız
def pisagor_ucgeni():
    pisagor_ucgeni_ucluleri = []
    for a_dik_kenari in range(1,101):
        for b_dik_kenari in range(1,101):
            hipotenüs = ((a_dik_kenari ** 2) + (b_dik_kenari ** 2)) ** 0.5
            if hipotenüs <= 100 and hipotenüs % 1 == 0:
                pisagor_uclusu = (a_dik_kenari, b_dik_kenari, hipotenüs)
                pisagor_ucgeni_ucluleri.append(pisagor_uclusu)
    return pisagor_ucgeni_ucluleri
pisagor_ucgeni()
````

# Problem 4

Bir duvar boyamaya karar verdiniz. Boya kutusunun üzerindeki talimatta, 1 kutu boyanın 5 metrekarelik bir duvarı boyayabileceği yazıyor. Rastgele bir duvar yüksekliği ve genişliği verildiğinde, kaç kutu boya satın almanız gerektiğini hesaplayın.

kutu sayısı = (duvar yüksekliği * duvar genişliği) ÷ kutu başına kaplama.

örneğin Yükseklik = 2, Genişlik = 4, Kaplama = 5

kutu sayısı = (2 * 4) ÷ 5 = 1.6

Ancak bir kutu boyanın 0,6'sını satın alamayacağınız için, sonuç 2 kutuya yuvarlanmalıdır .


````python
# 4.Problemin Çözümünü Buraya Yazınız
def boya_kutusu_sayisi():
    duvar_yuksekligi = int(input("Lütfen duvar yuksekliginin giriniz: "))
    duvar_genisligi = int(input("Lütfen duvar genisligini giriniz: "))
    kutu_sayisi = (duvar_genisligi * duvar_yuksekligi) / 5
    if kutu_sayisi % 1 != 0:
        kutu_sayisi = (kutu_sayisi // 1) + 1
        print(kutu_sayisi)
    elif kutu_sayisi % 1 == 0:
        print(kutu_sayisi)
boya_kutusu_sayisi()
````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
# 5.Problemin Çözümünü Buraya Yazınız
def harf_sayilari():
    unlu_harfler = ["a","e","i","ı","u","ü","o","ö"]
    unsuz_harfler = ["b","c","ç","d","f","g","ğ","h","j","k","l","m","n","p","r","s","ş","t","y","v","z"]
    kelime = input("Lütfen kelime giriniz: ")
    unlu_harf_sayisi = 0
    unsuz_harf_sayisi = 0
    for i in unlu_harfler:
        if i in kelime:
            unlu_sayi = kelime.count(i)
            unlu_harf_sayisi += unlu_sayi
    for i in unsuz_harfler:
        if i in kelime:
            unsuz_sayi = kelime.count(i)
            unsuz_harf_sayisi += unsuz_sayi
    print(f"{kelime} kelimesi {unlu_harf_sayisi} ünlü harf, {unsuz_harf_sayisi} ünsüz harf bulundurur.")
harf_sayilari()
````

# Problem 6

Verilen bir listeden çift sayıları yazdıran bir fonksiyon yazınız.


````python
# 6.Problemin Çözümünü Buraya Yazınız
sayilar = []
while True:
    sayi = input("Lütfen sayilari girin: ")
    if sayi == "exit":
        break
    sayi_int = int(sayi)
    sayilar.append(sayi_int)
def cift_sayilar():
    cift_olan_sayilar = []
    for sayi in sayilar:
        if sayi % 2 == 0:
            cift_olan_sayilar.append(sayi)
    print(cift_olan_sayilar)
cift_sayilar()
````

# Problem 7

0'dan 10'a kadar olan sayıların toplamını hesaplayan özyinelemeli (Recursive Function) bir fonksiyon oluşturan bir fonksiyon yazınız.

Özyinelemeli (Recursive Function) bir işlev, kendini tekrar tekrar çağıran bir işlevdir.

````python
# 7.Problemin Çözümünü Buraya Yazınız
def sayilar_toplam(sayi):
    if sayi == 1:
        return 1
    elif sayi > 1:
        return sayi + sayilar_toplam(sayi - 1)
sayilar_toplam(10)
````
